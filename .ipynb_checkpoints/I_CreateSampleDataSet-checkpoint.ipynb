### Step 2:

The NSAF dataset is too large to clean in entirety due to the amount of "hand" editing that needs to be done and additional art historical context columns that will need to be added. To make sure the sample is well balanced, I will make a stratified sample based on the "Category" attribute. This is also one of the data-points most consistently included in the original NSAF website data.

```python
## describing category variable
df1['Category'].describe()
```

```python
## checking category distribution
percentage_counts = df1['Category'].value_counts(normalize=True) * 100
print(percentage_counts)
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Nixie One','Garamond', 'Times New Roman']
```

```python
top10 = percentage_counts.nlargest(10)
top10.plot.pie(figsize=(10, 5))

plt.ylabel('')
plt.title('NSAF OBJECT CATEGORIES', color="#471203", fontweight='bold', fontsize=20, pad=10)
plt.show()
```

```python
prim_green = "#006F62"
soft_green = "#D1E0D7"
peach = "#F8CFA9"
almond = "#ECE1D5"
```

```python
data0 = {'cat':['Books and Ephemera','Drawings and Watercolors','Paintings','Icons, Triptychs, and Diptychs','Photographs','Prints','Sculptures','Bowls','Ceramic Tea aand Coffee Wares',
                'Clothing and Costumes','Coins and Paper Money','Collectibles','Clocks and Timepieces','Dollas and Figurines','Ethnographic Works of Art',
                'Firearms and Blades', 'Flatware','Holloware', 'Instruments','Jewelry','Lamps and Lighting',
                'Ornamental Ceramic Wares', 'Other - Altars','Other - Arts and Antiques','Other - Assorted','Textiles','Vases'],
         'per':[6.01,4.41,35.82,0.44,1.87,11.20,16.14,0.9,0.31,0.33,1.34,1.27,0.78,1.0,0.59,0.37,0.22,0.29,0.61,3.62,0.46,1.86,0.26,5.06,2.35,1.4,1.12]}
data0 = pd.DataFrame(data0)
```

```python
sns.set_style(rc={"axes.facecolor": "white", "figure.facecolor": "#F5F1E4", 'axes.edgecolor':'#471203',
                 'xtick.color':'#523A34','xtick.labelsize': 12,'ytick.color':'#523A34','font.family':'serif'})

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

# Modify existed axis
ax.set_aspect(aspect="equal")
ax.set_title('NSAF Condensed Categories Waffle', y=1.05, pad=5, fontsize=20, color="#471203", fontweight='bold')

Waffle.make_waffle(
    ax=ax,
    rows=10,
    columns=14,
    values=data0['per'],
    colors=[prim_green,"#F2FF00","#F800F8",soft_green,soft_green,soft_green,
                  soft_green,almond,almond,almond,almond,almond,
                  almond,almond,almond,almond,almond,almond,
                  almond,almond,almond,almond,almond,almond,
                  almond,almond,almond],
    labels=list(data0['cat']),  # Legend would not be created without this line
    legend={'loc': 'upper left', 'bbox_to_anchor': (-.4,1.03)}
)

plt.show()
fig.savefig("category_waffle", dpi=300)
```

```python
## making sample
fbi_messy_sample = df1.groupby('Category', group_keys=False).apply(lambda x: x.sample(frac=0.1))
```

```python
##  confirming sample has accurate distribution
sample_percentage_counts = fbi_messy_sample['Category'].value_counts(normalize=True) * 100
print(sample_percentage_counts)
```

```python
## final check on sample size etc.
fbi_messy_sample.info()
```
